Let's create the Elliptic Curve necessary to carry out the rest of the assignment :

```python
field = PolynomialRing(Zmod(sqrt(Q1_q)), 'x')
x = field.gens()[0]
polynomial = x ** 2 + 1
quotient = field.quotient(polynomial, 'x')
x = quotient.gens()[0]
E = EllipticCurve(quotient, Q1_E)
```
